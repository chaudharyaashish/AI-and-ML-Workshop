```python
import numpy as np
import matplotlib . pyplot as plt
from PIL import Image
# Load image and convert to grayscale
image = Image.open ("/content/drive/MyDrive/Week1 AI/cameraman.png") . convert ("L")
# Convert to NumPy array
image_array = np . array ( image )
# Get image dimensions
height , width = image_array . shape
# Reshape the image into a 2D array where each row represents a row of pixels
# This allows PCA to process each row as an observation and find patterns across pixel intensities
data = image_array . reshape ( height , width )
display(data)
```

```python
# Compute the mean of each column ( feature )
mean = np . mean ( data , axis =0)
# Subtract mean to center the data
centered_data = data - mean
display(centered_data)
```

```python
cov_matrix = np . cov ( centered_data , rowvar = False )
display(cov_matrix)
```

```python
# Compute eigenvalues and eigenvectors
eigenvalues , eigenvectors = np . linalg . eigh ( cov_matrix )
# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np . argsort ( eigenvalues ) [:: -1]
eigenvalues = eigenvalues [ sorted_indices ]
eigenvectors = eigenvectors [: , sorted_indices ]
display(eigenvalues)
display(eigenvectors)
```

```python
explained_variance_ratio = eigenvalues / np .sum ( eigenvalues )
plt . plot ( np . cumsum ( explained_variance_ratio ) )
plt . title (" Cumulative Explained Variance ")
plt . xlabel (" Number of Components ")
plt . ylabel (" Cumulative Explained Variance ")
plt . grid ( True )
plt . show ()
```

```python
k = 50 # Choose k principal components
components = eigenvectors [: , : k ]
display(components)
```

```python
compressed_data = np . dot ( centered_data , components )
display(compressed_data)
```

```python
decompressed_data = np . dot ( compressed_data , components . T ) +mean
display(decompressed_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load and convert the image to grayscale
image = Image.open("/content/drive/MyDrive/Week1 AI/cameraman.png").convert("L")
image_array = np.array(image)  # Convert to NumPy array

# Function to apply Singular Value Decomposition (SVD) compression
def compress_image(img_array, k):
    """Compresses an image using SVD by keeping only 'k' singular values."""
    U, S, Vt = np.linalg.svd(img_array, full_matrices=False)
    S = np.diag(S[:k])  # Keep only top k singular values
    compressed = U[:, :k] @ S @ Vt[:k, :]  # Reconstruct image with k components
    return compressed

# Set k value (compression level)
k = 50  # Adjust k to change compression quality

# Compress and reconstruct the image
compressed_data = compress_image(image_array, k)
decompressed_data = np.clip(compressed_data, 0, 255).astype(np.uint8)  # Ensure valid pixel values

# Plot the results
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(image_array, cmap="gray")
plt.title("Original Image")
plt.axis("off")

# Compressed Representation
plt.subplot(1, 3, 2)
plt.imshow(compressed_data, cmap="gray", aspect="auto")
plt.title(f"Compressed Image (k={k})")
plt.axis("off")

# Decompressed Image
plt.subplot(1, 3, 3)
plt.imshow(decompressed_data, cmap="gray")
plt.title("Decompressed Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```
